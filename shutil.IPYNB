```python
#此為將A檔案內容複製到B檔案中(檔案產生時間會是新的) shutil.copyfileobj
import os
import shutil
#os.chdir('C:/Vscode_DATA/move_file/TESTFILE1')  # 更改工作目錄，有點像cd指令
f1 = open('C:/Vscode_DATA/move_file/TESTFILE1/A.txt','r')     # 開啟為可讀取
f2 = open('C:/Vscode_DATA/move_file/TESTFILE1/B.txt','a')    # 開啟為可添加
shutil.copyfileobj(f1,f2)           # 複製內容

```

```python
#此為將A檔案取代性複製到B檔案中(檔案產生時間會是新的) shutil.copyfile
import os
import shutil
#os.chdir('C:/Vscode_DATA/move_file/TESTFILE1')  # 更改工作目錄，有點像cd指令
f1 = 'C:/Vscode_DATA/move_file/TESTFILE1/A.txt'    # 欲複製的檔案
f2 = 'C:/Vscode_DATA/move_file/TESTFILE2/AA.txt'  # 存檔的位置與檔案名稱
shutil.copyfile(f1,f2)   # 複製檔案
```

```python
#此為複製檔案權限資訊 shutil.copymode
import shutil
f1 = 'C:/Vscode_DATA/move_file/TESTFILE1/A.txt'
f2 = 'C:/Vscode_DATA/move_file/TESTFILE1/B.txt' 
shutil.copymode(f1,f2)   # 複製權限資訊
```

```python
#此為複製檔案(包含權限資訊)到指定檔案 shutil.copy
import shutil
f1 = 'C:/Vscode_DATA/move_file/TESTFILE1/A.txt'
f2 = 'C:/Vscode_DATA/move_file/TESTFILE1/NWE.txt' 
shutil.copy(f1,f2)     # 將 demo/demo.txt 複製到 demo2 資料夾的 demo.txt
```

```python
#此為複製檔案的所有資訊(權限資訊、修改時間...等)到指定檔案 shutil.copystat
import shutil
f1 = 'C:/Vscode_DATA/move_file/TESTFILE1/A.txt'
f2 = 'C:/Vscode_DATA/move_file/TESTFILE1/B.txt' 
shutil.copystat(f1,f2)   # 複製權限資訊、修改時間...等資訊
```

```python
#此為複製檔案及所有資訊到指定位置 shutil.copy2
#基本上使用這個指令不會有檔案產生時間差異
import shutil
f1 = 'C:/Vscode_DATA/move_file/TESTFILE1/A.txt'
f2 = 'C:/Vscode_DATA/move_file/TESTFILE2/AA.txt' 
shutil.copy2(f1,f2)     
```

```python
#此為移動(取代)檔案 shutil.move
import shutil
f1 = 'C:/Vscode_DATA/move_file/TESTFILE1/A.txt'
f2 = 'C:/Vscode_DATA/move_file/TESTFILE2/AA.txt' 
shutil.move(f1,f2)
```

```python
#此為複製整個資料夾及內容到指定位置(無法取代) shutil.copytree
#src	來源資料夾
#dst	目標資料夾
#ignore	要忽略的檔案，使用 shutil.ignore_patterns，預設 None
#copy_function	複製模式，預設 shutil.copy2
#ignore_dangling_symlinks	是否屏蔽符號鏈接錯誤，預設 False
#symlinks	是否屏蔽不存在路徑的錯誤，預設 False
#dirs_exist_ok     目標資料夾已存在時是否允許複製，預設 False (Python 3.8 新增)
import shutil
f1 = 'C:/Vscode_DATA/move_file/TESTFILE3/001/'
f2 = 'C:/Vscode_DATA/move_file/TESTFILE4/001/' 
#shutil.copytree(f1, f2, ignore=shutil.ignore_patterns('*.jpg', '*.png'))
shutil.copytree(f1, f2,copy_function=shutil.copy2 ,dirs_exist_ok=True)  
```

```python
#此為刪除整個資料夾跟檔案 shutil.rmtree
import shutil
f1 = 'C:/Vscode_DATA/move_file/TESTFILE4/001/' 
shutil.rmtree(f1)
```

```python
#此為將整個資料夾壓縮成zip檔案 shutil.make_archive
#base_name	壓縮後的檔案名稱 ( 可以使用目錄 + 名稱 )。
#format	壓縮格式，可使用 zip、tar...等。
#base_dir	相對於根目錄的目錄。
#root_dir	欲壓縮的檔案根目錄，預設為執行程式的目錄。
#owner	檔案擁有者，系統預設。
#group	檔案群，系統預設。
#logger	記錄日誌，預設 logging.Logger 對象。
import shutil
f1 = 'C:/Vscode_DATA/move_file/arch' 
f2 = 'C:/Vscode_DATA/move_file/TESTFILE4/' 
shutil.make_archive(f1, 'zip', base_dir="001", root_dir=f2)
```

```python
#此為解壓縮zip檔案到指定資料夾 shutil.unpack_archive
#file	欲解壓縮的檔案 ( 路徑相對於程式執行的目錄 )。
#format	解壓縮格式，預設以檔案的副檔名為主。
#extract_dir	解壓縮之後放置的檔案目錄，預設為程式執行的目錄。
import shutil
f1 = 'C:/Vscode_DATA/move_file/arch.zip' 
f2 = 'C:/Vscode_DATA/move_file/unzip'
shutil.unpack_archive(f1,f2)
```
